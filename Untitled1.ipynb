```python
if 'annotations' in metadata:
    annotations = metadata['annotations']
    
    # Debugging: Check the type and length of annotations
    print(f"Type of annotations: {type(annotations)}")
    print(f"Number of annotations: {len(annotations)}")

    # Check if annotations is a list and contains items
    if isinstance(annotations, list) and len(annotations) > 0:
        # Check each annotation and its structure
        expected_keys = ['image_id', 'category_id', 'bbox', 'area', 'iscrowd', 'ignore', 'id']
        for i, annotation in enumerate(annotations):
            is_valid = True  # To track if any issue is found with this annotation
            
            # Debugging: Print the first annotation's structure
            if i == 0:
                print(f"First annotation structure: {annotation}")
            
            # Check for missing or unexpected fields
            for key in expected_keys:
                if key not in annotation:
                    print(f"⚠️ Missing key '{key}' in annotation {i}")
                    is_valid = False
                elif key == 'bbox':
                    # Check if 'bbox' is a list with 4 numbers
                    bbox = annotation['bbox']
                    if not isinstance(bbox, list) or len(bbox) != 4 or not all(isinstance(x, (int, float)) for x in bbox):
                        print(f"⚠️ Invalid 'bbox' structure in annotation {i}: {bbox}")
                        is_valid = False

            # If any strange issue is found, print the problematic annotation
            if not is_valid:
                print(f"⚠️ Found a strange annotation at index {i}: {annotation}")
    else:
        print("⚠️ The 'annotations' key does not contain a list or is empty.")
else:
    print("⚠️ 'annotations' key not found in metadata.")
```

```python
if 'annotations' in metadata:
    annotations = metadata['annotations']
    
    # List to store all keys across all annotations
    all_keys = []

    # Loop through annotations and collect keys
    for i, annotation in enumerate(annotations):
        # List the keys of the current annotation
        annotation_keys = list(annotation.keys())
        all_keys.extend(annotation_keys)  # Add them to the all_keys list

        # Optionally print random annotation and its keys
        if i == random.randint(0, len(annotations) - 1):  # Select a random index
            print(f"Random Annotation {i} Keys: {annotation_keys}")

    # Count the occurrence of each key
    key_counts = Counter(all_keys)

    # Print the count for each key
    print("\nTotal number of occurrences for each key:")
    for key, count in key_counts.items():
        print(f"Key: {key}, Count: {count}")
else:
    print("⚠️ 'annotations' key not found in metadata.")
```

```python
import json
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import os

# Load metadata
metadata_path = "dataset/updated_train.json"  # Update with your actual path
with open(metadata_path, "r") as f:
    metadata = json.load(f)


# Check if 'annotations' key exists
if 'annotations' in metadata:
    annotations = metadata['annotations']

    # Choose the image_id (replace with the name of the image you want to display)
    image_id_to_display = "00345.jpg"  # Update with the specific image name you want

    # Find the annotation corresponding to the image_id
    selected_annotation = None
    for annotation in annotations:
        if annotation['image_id'] == image_id_to_display:
            selected_annotation = annotation
            break

    if selected_annotation is not None:
        # Get the image name and the corresponding image path
        image_name = selected_annotation['image_id']
        image_path = os.path.join("dataset/updated_train", image_name)  # Update with the actual image folder path
        
        # Check if the image exists
        if os.path.exists(image_path):
            # Open the image
            img = Image.open(image_path)
            
            image_size = img.size  # (width, height)
            print(f"Image Size: {image_size[0]} x {image_size[1]}")
            # Prepare metadata information
            category_id = selected_annotation['category_id']
            bbox = selected_annotation['bbox']  # [x_min, y_min, width, height]
            area = selected_annotation['area']
            iscrowd = selected_annotation['iscrowd']
            ignore = selected_annotation['ignore']
            annotation_id = selected_annotation['id']

            # Set up the plot
            fig, ax = plt.subplots(1, figsize=(10, 10))
            ax.imshow(img)
            
            # Add the bounding box
            x_min, y_min, width, height = bbox
            rect = patches.Rectangle((x_min, y_min), width, height, linewidth=2, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

            # Display metadata information
            metadata_text = f"Category ID: {category_id}\nArea: {area}\nIs Crowd: {iscrowd}\nIgnore: {ignore}\nAnnotation ID: {annotation_id}"
            ax.text(10, 10, metadata_text, fontsize=12, color='white', bbox=dict(facecolor='black', alpha=0.5))

            # Show the image with the bounding box and metadata
            plt.title(f"Image: {image_name}")
            plt.axis('off')  # Hide axes
            plt.show()

        else:
            print(f"⚠️ Image {image_name} not found at {image_path}")
    else:
        print(f"⚠️ No annotation found for image {image_id_to_display}")
else:
    print("⚠️ 'annotations' key not found in metadata.")
```
