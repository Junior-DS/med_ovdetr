```python
import pickle
import matplotlib.pyplot as plt
import cv2

def visualize_proposals(image_path, proposal_path):
    # Load image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Load proposals
    with open(proposal_path, 'rb') as f:
        proposals = pickle.load(f)
    
    # Debug: Print keys and sample values
    print("Proposal Keys:", proposals.keys())
    
    # Check if the image name exists in proposals
    image_key = image_path.split('/')[-1]  # Extract filename
    if image_key not in proposals:
        print(f"Warning: {image_key} not found in proposals!")
        return

    boxes = proposals[image_key]['boxes']  # Get boxes for the image
    
    # Debug: Print first few bounding boxes
    print(f"Total Proposals: {len(boxes)}")
    print("First 5 Boxes:", boxes[:5])

    # Visualize
    plt.figure(figsize=(12, 8))
    plt.imshow(img)
    ax = plt.gca()
    
    for box in boxes[:10]:  # Draw first 10 proposals
        x1, y1, x2, y2 = box
        ax.add_patch(plt.Rectangle(
            (x1, y1), x2 - x1, y2 - y1,
            fill=False, edgecolor='red', linewidth=1
        ))
    
    plt.title(f"Proposals for {image_key}")
    plt.show()

# Example usage
visualize_proposals(
    "dataset/updated_train/00017.jpg",
    "dataset/proposals/base_train_proposals.pkl"
)
```

```python
import pickle

def find_max_proposals(proposal_path):
    # Load proposals
    with open(proposal_path, 'rb') as f:
        proposals = pickle.load(f)

    max_proposals = 0
    max_image = None  # To track which image has the most proposals

    for image_name, proposal_data in proposals.items():
        num_proposals = len(proposal_data['boxes'])  # Count number of proposals
        if num_proposals > max_proposals:
            max_proposals = num_proposals
            max_image = image_name

    print(f"Max Proposals: {max_proposals} in {max_image}")

# Example usage
find_max_proposals("dataset/proposals/base_train_proposals.pkl")
```

```python
import pickle
from collections import Counter

def count_proposals(proposal_path):
    # Load proposals
    with open(proposal_path, 'rb') as f:
        proposals = pickle.load(f)

    # Count occurrences of different proposal counts
    proposal_counts = Counter(len(proposal_data['boxes']) for proposal_data in proposals.values())

    # Print the results
    for num_proposals, num_images in sorted(proposal_counts.items()):
        print(f"Images with {num_proposals} proposals: {num_images}")

# Example usage
count_proposals("dataset/proposals/novel_train_proposals.pkl")
```
